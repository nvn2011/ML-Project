```r
kmo = function(x)
    {
    x=subset(x,complete.cases(x))
    r=cor(x)
    r2=r^2
    i=solve(r)
    d=diag(i)
    p2=(-i/sqrt(outer(d,d)))^2
    diag(r2)=diag(p2)=0
    KMO=sum(r2)/(sum(r2)+sum(p2))
    #MSA=colSums(r2)/(colSums(r2)+colSums(p2))
    return(list(KMO=KMO))
}
```
